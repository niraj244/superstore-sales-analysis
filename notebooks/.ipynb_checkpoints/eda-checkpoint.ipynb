## Day 1 Exploration Questions
Finding out answers of 7 basic questions about the dataset before any analysis.

```python
# How many orders are in this dataset? and how many of them are unique?
print(f"Total rows: {len(df)}")
print(f"Total Unique orders: {df['Order ID'].nunique()}")
```

```python
# Q2: What time period does this data cover?
order_time= pd.to_datetime(df['Order Date'])
print(f"The order date ranges from {order_time.min().date()} to {order_time.max().date()}")
```

```python
# Q3: How many product categories do you have?
print(f" Total number of product categories: {df['Category'].value_counts()}")
```

```python
#Q4: Who are our customers(segments)?
print(f"Customer Segments: {df['Segment'].value_counts()}")
```

```python
#Q5: What regions are the customers from?
print(f" Customer regions: {df['Region'].value_counts()}")
```

```python
# Q6: What is the total sales and total profit in the dataset?
print(f" Total sales: {df['Sales'].sum()} ,and \n Total profit: {df['Profit'].sum()}")
```

```python
#Q7: What is the overall profit margin?
margin= (df['Profit'].sum()/df['Sales'].sum())*100
print(f"Total profit margin percentage: {margin:.2f}%")
```
