```python
## Day 2: Data Cleaning
Goal: Fix data types, add useful columns, and document all decisions.
```

```python
print(df['Order Date'].dtype)
print(df['Ship Date'].dtype)
print(df['Order Date'].head)
```

```python
df['Order Date']=pd.to_datetime(df['Order Date'])
df['Ship Date']=pd.to_datetime(df['Ship Date'])
print(df['Order Date'].dtype)
print(df['Ship Date'].dtype)
```

```python
df['Order Month']=df['Order Date'].dt.month
df['Order Year']=df['Order Date'].dt.year
print(df[['Order Date','Order Month', 'Order Year']].head())
```

```python
df['Order Month Name']=df['Order Date'].dt.strftime('%b')
print(df[['Order Date', 'Order Month','Order Month Name']].head(10))
```

```python
#Days to Ship?
#Average days needed to ship the product
#Max days to ship
#min days to ship

df['Days to Ship']=(df['Ship Date']-df['Order Date']).dt.days
print( "Ship days:\n", df['Days to Ship'].head(10))
print(f"Average days to ship: {df['Days to Ship'].mean():.2f} days")
print(f"Max days to ship: {df['Days to Ship'].max()} days")
print(f"Min days to ship: {df['Days to Ship'].min()} days")
```

```python
#Profit margin
df['Profit Margin']=(df['Profit']/df['Sales'])*100
print(df[['Profit', 'Sales', 'Profit Margin']].head(5))
print(f"Average profit margin per row: {df['Profit Margin'].mean():.2f} %")
```
